# Data Preprocessing and cleaning


**Function to extract numbers from a list**

```python
import ast
def extract_numbers(lst_str):
    lst = ast.literal_eval(lst_str)
    return [float(num) for num in lst]
```

```python
# Extract features from the lists using apply()
data['prev_9_num_1'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[0])
data['prev_9_num_2'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[1])
data['prev_9_num_3'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[2])
data['prev_9_num_4'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[3])
data['prev_9_num_5'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[4])
data['prev_9_num_6'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[5])
data['prev_9_num_7'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[6])
data['prev_9_num_8'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[7])
data['prev_9_num_9'] = data['previous 9 number'].apply(lambda x: extract_numbers(x)[8])
```

```python
# ... continue extracting other elements from the lists

data['prev_8_pip_change_1'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[0])
data['prev_8_pip_change_2'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[1])
data['prev_8_pip_change_3'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[2])
data['prev_8_pip_change_4'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[3])
data['prev_8_pip_change_5'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[4])
data['prev_8_pip_change_6'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[5])
data['prev_8_pip_change_7'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[6])
data['prev_8_pip_change_8'] = data['prev 8 pip change'].apply(lambda x: extract_numbers(x)[7])
```

```python
# ... continue extracting other elements from the lists

data['prev9_prices_1'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[0])
data['prev9_prices_2'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[1])
data['prev9_prices_3'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[2])
data['prev9_prices_4'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[3])
data['prev9_prices_5'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[4])
data['prev9_prices_6'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[5])
data['prev9_prices_7'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[6])
data['prev9_prices_8'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[7])
data['prev9_prices_9'] = data['prev9 prices'].apply(lambda x: extract_numbers(x)[8])
```

Specifically, the columns with names 'previous 9 number', 'prev 8 pip change', and 'prev9 prices' are removed from the DataFrame, likely because these columns contain lists of data or are no longer required for the subsequent analysis.

**Selecting Columns**: A list named selected_columns is created, containing the names of specific columns from the original dataset that need to be used in the analysis. These columns are selected based on their relevance to the analysis.

**Converting Data Types**: The data DataFrame (assuming it contains the original dataset) is modified using the astype(float) method. This operation converts the data in the selected columns from their original data types (e.g., strings or integers) to the floating-point data type (float). Converting the data to float allows for numerical computations and analyses.

By performing these steps, the data is being prepared for further analysis, ensuring that it contains only the relevant numerical data in the selected columns and is in a suitable data type for calculations and modeling.
